```python
# 【要生成】2D シミュレーション .npy ファイルが必要
# project2 の surface_wave_2d_2/ に相当する kusabi 用 2D シミュレーション結果
filename  = r"C:\Users\cs16\Roughness\project4\surface_wave_2d\Kusabi\T1_series_kusabi_pitch125_depth20.npy"
filename2 = r"C:\Users\cs16\Roughness\project4\surface_wave_2d\Kusabi\T3_series_kusabi_pitch125_depth20.npy"
data_kabe_T1 = np.load(filename)
data_yuka_T3 = np.load(filename2)

z1 = 49
z2 = 224
x1 = 30
x2 = 40
dt_ns = 0.71  # ns per timestep
time_range = sourse_new_2.right + sourse_new_2.left
# プローブ発射を0とした実時間軸
start_time_step = 4000  # 描画開始時のタイムステップ
time_axis_us = [(start_time_step + i) for i in range(time_range)]

fig = plt.figure()
im = plt.imshow(
    -data_kabe_T1[x1:x2, z1, 1214 - sourse_new_2.left : 1214 + sourse_new_2.right],
    extent=(time_axis_us[0], time_axis_us[-1], 0, 0.10),
    interpolation="nearest",
    cmap="bwr",
    aspect="auto",
)
plt.xlabel("Time [\u03bcs]")
plt.ylabel(r"x-axis position [mm]")

ax = plt.gca()
current_yticks = list(ax.get_yticks())
if 0.10 not in current_yticks:
    current_yticks.append(0.10)
    ax.set_yticks(sorted(current_yticks))

plt.text(0.02, 0.1, "(a)", transform=plt.gca().transAxes, fontsize=25, weight="bold", ha="left", va="top")
cbar = plt.colorbar(im)
cbar.set_label("[Pa]", rotation=0, labelpad=10)
# plt.savefig(r"C:\Users\cs16\Roughness\project4\tmp_output\kusabi_Figure1000a.eps", format="eps", dpi=300)
plt.show()

plt.rcParams["font.size"] = 13
fig = plt.figure()
im = plt.imshow(
    -data_yuka_T3[x2 - 2, z1:z2, 1214 - sourse_new_2.left : 1214 + sourse_new_2.right],
    extent=(time_axis_us[0], time_axis_us[-1], 0, 1.75),
    interpolation="nearest",
    cmap="bwr",
    aspect="auto",
)
plt.xlabel("Time [\u03bcs]")
plt.ylabel(r"z-axis position [mm]")

ax = plt.gca()
current_yticks = list(ax.get_yticks())
if 1.75 not in current_yticks:
    current_yticks = [tick for tick in current_yticks if abs(tick - 1.75) > 0.01]
    current_yticks.append(1.75)
    ax.set_yticks(sorted(current_yticks))

ax.set_ylim(0, 1.75)
plt.text(0.02, 0.1, "(b)", transform=plt.gca().transAxes, fontsize=25, weight="bold", ha="left", va="top")
cbar = plt.colorbar(im)
cbar.set_label("[Pa]", rotation=0, labelpad=10)
# plt.savefig(r"C:\Users\cs16\Roughness\project4\tmp_output\kusabi_Figure1000b.eps", format="eps", dpi=300)
plt.show()
```
